```python
import numpy as np
import matplotlib.pyplot as plt

def compute_integral_trapez(x, y, nbi):
    """
    Job     : Function to approximate surcafe of curve using trapez method
    Input   : X and Y point, Nbi: nbr of point
    Output  : Value of Integral
    """
    integral = 0
    for i in range(nbi):
        integral += ((y[i] + y[i+1]) * (x[i+1] - x[i]) * 0.5)
    return integral

def plot_integral(x, y, nbi):
    """
    Job     : Function to Draw each calculated trapez area as approximation is going
    Input   : X and Y point, Nbi: nbr of point
    Output  : None
    """
    for i in range(nbi):
        posx = [x[i], x[i], x[i+1], x[i+1], x[i]]
        posy = [0, y[i], y[i+1], 0, 0]
        plt.plot(posx, posy, 'r')
```

```python
xmin = 0
xmax = 3*np.pi/2
nbx = 20
nbi = nbx - 1 # nombre d'intervalles

x = np.linspace(xmin, xmax, nbx)
y = np.cos(x) 

plt.plot(x,y,"bo-")
integral = compute_integral_trapez(x, y, nbi)
plot_integral(x, y, nbi)   
print("integrale =", integral)

plt.show()
```
